# raglet — hybrid vs. dense-only retrieval

raglet fuses dense (semantic) and sparse (BM25) signals with Reciprocal Rank Fusion. This notebook compares the two modes on the same query.

```python
from raglet import RAG, RAGConfig

def top_sources(rag, query, sparse):
    rag.config.use_sparse = sparse
    return [c["source"] for c in rag.retrieve(query, k=3)]

rag = RAG(RAGConfig(embedder="hash", store_path="./.store"))
rag.ingest("./docs")

query = "how do embeddings map text?"
print("dense only :", top_sources(rag, query, sparse=False))
print("hybrid     :", top_sources(rag, query, sparse=True))
```
