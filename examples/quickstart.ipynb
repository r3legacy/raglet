# raglet — quickstart

A minimal end-to-end example. The default `hash` embedder and `extractive` LLM need **no model downloads**, so this runs immediately.

```python
from raglet import RAG, RAGConfig

# Point this at a folder of .txt / .md / .pdf / .docx files.
DOCS = "./docs"

rag = RAG(RAGConfig(
    embedder="hash",
    llm="extractive",
    store_path="./.store",
))
rag.ingest(DOCS)

result = rag.ask("What is the main feature described here?")
print(result["answer"])
```
